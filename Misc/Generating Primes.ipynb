```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sieve
```

## Generating Prime Numbers

```python
def get_primes(n):
    return np.array(list(sieve.primerange(1, n)))
```

```python
p = get_primes(102)
```

```python
# Difference between Primes n-1
dif = p[1:] - p[:-1]
d = np.fft.fft(dif)
```

```python
plt.figure(figsize=(20,10))
plt.plot(d, 'o')
```
